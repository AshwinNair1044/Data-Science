# Answer 1

```
#animated plot of cos graph
stop = 2*np.pi
fig, ax = plt.subplots()
x = np.arange(0, stop, 0.001)
y = np.sin(x)
l = plt.plot(x, y)
ax = plt.axis([0,stop,-1,1])
redDot, = plt.plot([0], [np.sin(0)], 'g+',)
def animate(i):
    redDot.set_data(i, np.sin(i))
    return redDot,
ani=FuncAnimation(fig, animate, frames=np.arange(0.0, stop, 0.1),interval=10, blit=True, repeat=True)
ani
```

```
df = pd.read_csv('/content/drive/My Drive/day.csv')
sns.set_style('white')
sns.set_style('ticks')
plt.figure(figsize=(7,5))
sns.scatterplot('temp','cnt',data=df,hue='season',palette=['purple','yellow','blue','green'])
plt.legend(bbox_to_anchor=(1.5,0.7))
plt.xlabel('Normalized Temperature', fontsize='large')
plt.ylabel('Count of Total Bike Rentals', fontsize='large');
```

# Answer 4 

```
plt.figure(figsize=(10,7))
plt.scatter('temp','cnt',data=df[df['season']==1],color='green',label='Spring')
plt.scatter('temp','cnt',data=df[df['season']==2],color='orange',label='Summer')
plt.scatter('temp','cnt',data=df[df['season']==3],color='brown',label='Fall/Autumn')
plt.scatter('temp','cnt',data=df[df['season']==4],color='blue',label='Winter')
plt.ylabel('Count of Total Bike Rentals',fontweight='bold',fontsize=14)
plt.xlabel('Normalized Temperature',fontweight='bold',fontsize=14)
plt.title('Bike Rentals at Different Temperatures \nBy Season',fontweight='bold',fontsize=18)
plt.legend(bbox_to_anchor=(1,0.8),title='Season')
plt.show()
```

```
fig = plt.figure(figsize=(15,10))
fontdict={'fontsize': 25,'weight' : 'bold'}
fontdicty={'fontsize': 18,'weight' : 'bold','verticalalignment': 'baseline','horizontalalignment': 'center'}
fontdictx={'fontsize': 18,'weight' : 'bold','horizontalalignment': 'center'}
plt.subplots_adjust(wspace=0.2, hspace=0.2)
fig.suptitle('Bike Rentals at Different Temperatures\nBy Season', fontsize=25,fontweight="bold", color="black",position=(0.5,1.01))
ax1 = fig.add_subplot(2,2,1)
ax1.scatter('temp', 'cnt', data=df[df['season']==1], c="green")
ax1.set_title('Spring', fontdict=fontdict, color="green")
ax1.set_ylabel("Count of Total Rental Bikes", fontdict=fontdicty, position=(0,-0.1))
#
ax2 = fig.add_subplot(2,2,2)
ax2.scatter('temp', 'cnt', data=df[df['season']==2], c="orange")
ax2.set_title('Summer', fontdict=fontdict, color="orange")
#
ax3 = fig.add_subplot(2,2,3)
ax3.scatter('temp', 'cnt', data=df[df['season']==3], c="brown")
ax3.set_title('Fall or Autumn', fontdict=fontdict, color="brown")
#
ax4 = fig.add_subplot(2,2,4)
ax4.scatter('temp', 'cnt', data=df[df['season']==4], c="blue")
ax4.set_title("Winter", fontdict=fontdict, color="blue")
ax4.set_xlabel("Normalized temperature", fontdict=fontdictx, position=(-0.1,0))
plt.show()
```

# Answer 8

```
cat = {"season":{1:"Spring",2:"Summer",3:"Fall/Autumn",4:"Winter"}}
day = df.copy()
day.replace(cat,inplace=True)
plt.figure(figsize=(7,7))
fontdict={'fontsize': 18,'weight' : 'bold','horizontalalignment': 'center'}
sns.set_context('talk', font_scale=0.9)
sns.set_style('white')
sns.scatterplot(x='temp', y='cnt', hue='season', data=day, style='season',palette=['green','orange','brown','blue'], legend='full')
plt.legend(scatterpoints=1,bbox_to_anchor=(1, 0.7),loc=2,borderaxespad=1,ncol=1,fontsize=14)
plt.xlabel('Normalized Temperature', fontsize=16, fontweight='bold')
plt.ylabel('Count of Total Bike Rentals', fontsize=16, fontweight='bold')
plt.title('Bike Rentals at Different Temperatures\nBy Season', fontdict=fontdict, color="black",position=(0.5,1.1))
plt.show()
```

# Answer 9

```
sns.set_context('talk', font_scale=0.9) 
sns.set_style('ticks')
g = sns.relplot(x='temp', y='cnt', hue='season', data=day,palette=['green','orange','brown','blue'],col='season', col_wrap=4, legend=False,height=6, 
                aspect=0.6, style='season', sizes=(800,1000))
g.fig.suptitle('Bike Rentals at Different Temperatures\nBy Season' ,position=(0.5,1.1), fontweight='bold', size=18)
g.set_xlabels("Normalized Temperature",fontweight='bold', size=15,labelpad=10)
g.set_ylabels("Count of Total Bike Rentals",fontweight='bold', size=20,labelpad=10);
```
